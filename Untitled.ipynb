```python
import pandas as pd
import numpy as np
```

```python
%matplotlib inline
```

```python
origData = pd.read_csv("europeEconomic.csv", sep="\t")
```

```python
origData.head()
```

```python
from scipy.cluster.hierarchy import dendrogram, linkage
from matplotlib import pyplot as plt
X = pd.DataFrame(origData)
```

```python
y = X.Country
X = X.drop(labels=['Country'], axis=1)
```

```python
names = np.array(y)
```

```python
X.head()
```

```python
Z = linkage(X, 'single')
fig = plt.figure(figsize=(25, 10))
dn = dendrogram(Z, labels=names)
plt.show()
```

```python
Z = linkage(X, 'complete')
fig = plt.figure(figsize=(25, 10))
dn = dendrogram(Z, labels=names)
plt.show()
```

```python
Z = linkage(X, 'average')
fig = plt.figure(figsize=(25, 10))
dn = dendrogram(Z, labels=names)
plt.show()
```

```python
#elbow graph with the assistance of Ahmed Besbes: https://stackoverflow.com/questions/41540751/sklearn-kmeans-equivalent-of-elbow-method
from sklearn.cluster import KMeans
distorsions = []
for k in range(1, 10):
    kmeans = KMeans(n_clusters=k)
    kmeans.fit(X)
    distorsions.append(kmeans.inertia_)

fig = plt.figure(figsize=(15, 5))
plt.plot(range(1, 10), distorsions)
plt.grid(True)
plt.title('Elbow curve')
```

```python
np.random.seed(42)

model = KMeans(n_clusters=3)
model.fit(X)
data2D = model.transform(X)
```

```python
kmLabels = model.labels_
kmLabels
```

```python
#results['Country'] = names
#plt.scatter(results[0], results[0], c=labels.astype(np.float), edgecolor='k')
plt.figure(figsize=(20,10))
LABEL_COLOR_MAP = {0:'g', 1:'m', 2:'c'}

label_color = [LABEL_COLOR_MAP[l] for l in kmLabels]
plt.scatter(data2D[:,0], data2D[:,1], c=label_color)
for i in range (0,26):
    xy=(data2D[i][0],data2D[i][1])
    plt.annotate(names[i],xy)
plt.show() 
```
